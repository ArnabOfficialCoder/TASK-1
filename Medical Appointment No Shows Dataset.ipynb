```python
#drop rows with missing values in key columns like Gender, Age, Neighbourhood, SMS_received

df = df.dropna(subset =['Gender', 'Age', 'Neighbourhood', 'SMS_received'])
```

```python
#recheck for missing values

df.isnull().sum()
```

```python
# Standardize 'gender' and 'no-show' text

df['Gender'] = df['Gender'].str.upper().str.strip()
df['No-show'] = df['No-show'].str.lower().str.strip()
```

```python
# Check Standardization

print("Unique values after standardizing:")
print("Gender:", df['Gender'].unique())
print("No-show:", df['No-show'].unique())
```

```python
# Convert ScheduledDay and AppointmentDay to datetime and format

df['ScheduledDay'] = pd.to_datetime(df['ScheduledDay']).dt.strftime('%d-%m-%Y')
df['AppointmentDay'] = pd.to_datetime(df['AppointmentDay']).dt.strftime('%d-%m-%Y')
```

```python
# Check Conversion

print("After converting ScheduledDay:", df['ScheduledDay'].head())
print("After converting AppointmentDay:", df['AppointmentDay'].head())
```

```python
# Clean column names: lowercase, underscores

df.columns = (
    df.columns.str.strip()
    .str.lower()
    .str.replace(' ', '_')
    .str.replace('-', '_')
)

# Fix column typo
df.rename(columns={'handcap': 'handicap'}, inplace=True)
```

```python
# Check Cleaning

print("Column names after renaming:", df.columns.tolist())
```

```python
# Finxing Data Types 

# Ensure age is integer
df['age'] = df['age'].astype(int)

# Confirm data types
df.dtypes
```
